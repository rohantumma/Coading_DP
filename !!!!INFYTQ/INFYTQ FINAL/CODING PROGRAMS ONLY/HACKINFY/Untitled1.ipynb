```python
def compareTriplets(a, b):
    res=[]
    al=0
    bo=0
    al += a[0] > b[0];
    bo += b[0] > a[0];
    al += a[1] > b[1];
    bo += b[1] > a[1];
    al += a[2] > b[2];
    bo += b[2] > a[2];
    res.append(al)
    res.append(bo)
    return res

a=[1 ,2 ,3]
b=[1 ,2 ,3]
result = compareTriplets(a, b)
print(result)
```
